```python
import pickle
import numpy as np

direct = '/home/harryemeric/Project/cs231n_project_box_convolution/HE_experiments/pkl_files/'
file = 'ssd_lr0.001_wd8e-05_sub1.0train_loss.pkl'
file2 = 'full_box_ssd_lr0.001_wd0.0005_sub1.0train_loss.pkl'
file3 = 'ssd_lr0.001_wd8e-05_sub1.0valid_loss.pkl'
file4 = 'full_box_ssd_lr0.001_wd0.0005_sub1.0valid_loss.pkl'
```

```python
with open(direct + file, 'rb') as handle:
    train_loss = pickle.load(handle)
with open(direct + file2, 'rb') as handle:
    train_loss_box = pickle.load(handle)
with open(direct + file3, 'rb') as handle:
    val_loss = pickle.load(handle)
with open(direct + file4, 'rb') as handle:
    val_loss_box = pickle.load(handle)
```
